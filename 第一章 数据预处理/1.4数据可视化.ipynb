# 1.4 数据可视化 样例代码

### 导入所需要的包

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

### 直方图

```python
# 示例数据
data = [1, 2, 2, 3, 4, 5, 5, 5, 6, 7]

# 使用matplotlib进行画图
plt.hist(data, bins=5, alpha=0.7, color='blue')
plt.title('Histogram Example')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

### 箱线图

```python
# 示例数据
# 其中20是异常值
data = [1, 2, 2, 3, 4, 5, 5, 5, 6, 7, 20]

plt.boxplot(data)
plt.title('Boxplot')
plt.ylabel('Value')
plt.show()
```

### 散点图

```python
# 样例数据
x = [1, 2, 3, 4, 5]
y = [2, 3, 2, 4, 5]

plt.scatter(x, y)
plt.title('Scatter Plot')
plt.xlabel('X Value')
plt.ylabel('Y Value')
plt.show()
```

### 热力图

```python
# 假设样例数据如下
data = np.random.rand(10, 10)

plt.imshow(data, cmap='hot', interpolation='nearest')
plt.title('Heatmap')
plt.colorbar()
plt.show()
```

### 提琴图

```python
sns.violinplot(data)
plt.title('Violin Plot')
plt.ylabel('Value')
plt.show()
```

### 饼图

```python
labels = 'A', 'B', 'C', 'D'
sizes = [15, 30, 45, 10]

plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Pie Chart')
plt.show()
```

### 透视表

```python
data = {
    'Product': ['Apples', 'Oranges', 'Bananas', 'Apples', 'Oranges', 'Bananas'],
    'Sales': [5, 6, 7, 2, 4, 3],
    'Region': ['North', 'North', 'North', 'South', 'South', 'South']
}

df = pd.DataFrame(data)

pivot_table = pd.pivot_table(df, values='Sales', index='Product', columns='Region', aggfunc='sum')
print(pivot_table)
```
